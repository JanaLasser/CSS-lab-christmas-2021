```python
import pandas as pd
import matplotlib.pyplot as plt
import emoji_resources as er
from os.path import join
```

```python
emojis = er.emojis
languages = er.languages
```

```python
src = "../data/counts"
counts = pd.DataFrame(columns=["baseline"] + list(emojis.keys()))
for lang in languages:
    fname = f"counts_language-{lang}_2019-01-01-to-2021-11-28.csv"
    lang_counts = pd.read_csv(join(src, fname), parse_dates=["date"])
    lang_counts = lang_counts.drop(columns=["date"])
    counts = counts.append(lang_counts.sum(), ignore_index=True)
counts["language"] = languages
counts = counts[["language", "baseline"] + list(emojis.keys())]

counts = counts.rename(columns=emojis)
counts = counts.set_index("language")
```
